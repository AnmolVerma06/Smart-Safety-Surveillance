```python
model = YOLO("best.pt")

results = model.predict(source="1", show=True)
print(results)


def capture_from_webcam():
    cap = cv2.VideoCapture(0)  # Index 0 typically refers to the first webcam
    return cap

# Function to capture video from IP camera
def capture_from_ip_camera(ip_address, username=None, password=None):
    # Construct the URL for the IP camera stream
    url = f"rtsp://{username}:{password}@{ip_address}/video"
    cap = cv2.VideoCapture(url)
    return cap

def main():
    # Flag to indicate whether to use webcam or IP camera
    use_webcam = True

    # If you want to use an IP camera, set the following parameters
    ip_address = "192.168.92.153"
    username = "your_username"  # if authentication is required
    password = "your_password"  # if authentication is required

    if use_webcam:
        cap = capture_from_webcam()
    else:
        cap = capture_from_ip_camera(ip_address, username, password)

    if not cap.isOpened():
        print("Error: Failed to open the camera.")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Failed to capture frame.")
            break

        # Display the frame
        cv2.imshow('Frame', frame)

        # Check for user input to switch between webcam and IP camera
        key = cv2.waitKey(1)
        if key == ord('q'):
            break
        elif key == ord('s'):  # Switch to IP camera
            use_webcam = False
            cap.release()
            cap = capture_from_ip_camera(ip_address, username, password)
        elif key == ord('w'):  # Switch to webcam
            use_webcam = True
            cap.release()
            cap = capture_from_webcam()

    # Release the camera and close all OpenCV windows
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
